```python
import pandas as pd
# Load the dataset to inspect its contents
data = pd.read_csv("C:/Users/kuppu/Downloads/Dataset .csv")
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Split the 'Cuisines' column into individual cuisines and count occurrences
cuisine_counts = data['Cuisines'].str.split(', ').explode().value_counts()

# Get the top 3 most common cuisines
top_cuisines = cuisine_counts.head(3)

# Calculate the percentage of restaurants serving each of the top cuisines
total_restaurants = len(data)
top_cuisines_percentages = (top_cuisines / total_restaurants) * 100

# Display the top 3 cuisines and their percentages
top_cuisines, top_cuisines_percentages
```
